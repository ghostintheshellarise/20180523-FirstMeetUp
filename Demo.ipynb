# Demonstration of using Python with building data

In this demonstration, we will

- import and preprocess large amount of data using Python Data Analysis Library (pandas)
- use pandas to extract data based on date and time
- calculate average building power consumption
- visualize the data
- categorize the data by clustering

## Importing and preprocessing data

Let's import the required pandas module 1st

```python
import pandas as pd
```

and we can read the data into a _pandas dataframe_ object from a file

```python
df = pd.read_excel('DemoFile.xlsx')
```

```python
df
```

Looks like each sheet has one type of data only. Can we read the data from another worksheet?

```python
df2 = pd.read_excel('DemoFile.xlsx', sheet_name='Sheet2')
```

```python
df2
```

Now we have two dataframes of data. Can we combine them into one for simplicity? __Yes!__

```python
df_final = df.merge(df2)
```

```python
df_final
```

The dataframes can be combined easily.

## Select data based on date and time

Let's try selecting some data based on date and time only. But will it be very difficult with date and time stored in text in the original file? __No!__

First, we need a function called _datetime_ in the _datetime_ module in Python

```python
from datetime import datetime
```

Now, let's see what it can do

```python
datetime(2015, 1, 2, 16, 35, 2)
```

But it can do even more. For example, if I only want data in 2016 from the dataframe, all I need is

```python
df_select = df_final.loc[(
    df_final.Timestamp >= datetime(2016, 1, 1)
) & (
    df_final.Timestamp < datetime(2017, 1, 1)
), :]
```

```python
df_select
```

And the technique is similar if you want data based on temperature. If you want data which temperature is between 28C and 30C only, you just need

```python
temp_col_name = '5509_Outdoor Temperature_TL Archive'
df_select2 = df_select.loc[((df_select[temp_col_name] >= 28) & (df_select[temp_col_name] <= 30)), :]
```

```python
df_select2
```

## Calculating mean building load

Want to know the mean building load within the selected data? All you need is

```python
df_select2.loc[:, '5521_Building Load_TL Archive'].mean()
```

## Data visualization

It's difficult to see what's going on with all the data, right? For example, how can we know the load pattern in July 2016? First, let's extract the data

```python
df_select3 = df_final.loc[(
    df_final.Timestamp < datetime(2016, 8, 1)
) & (
    df_final.Timestamp >= datetime(2016, 7, 1)
), :]
```

```python
df_select3
```

```python
%matplotlib inline
df_select3.plot(x='Timestamp', y='5521_Building Load_TL Archive')
```

The plot is confusing, right? Let's organize the timestamp first and visualize the data again. We create a new column using the time of the timestamp only.

```python
from datetime import time
```

```python
df_select3.loc[:, 'Time'] = [ts.hour for ts in df_select3.Timestamp]
```

```python
df_select3
```

And let's make a box plot using the new time column and the building load column

```python
import matplotlib.pyplot as plt
df_select3.boxplot(by='Time', column='5521_Building Load_TL Archive')
```

## Clustering

Can we understand the pattern more by using clustering algorithm (a machine learning algorithm to categorize data) with __both__ temperature and building load data? Let's try that with the _K-means_ algorithm from the _scikit-learn_ library to categorize the selected data into 5 groups.

```python
from sklearn.cluster import KMeans
df_select3.loc[:, 'Class'] = KMeans(n_clusters=4).fit_predict(df_select3.loc[:, [
    'Time', '5521_Building Load_TL Archive', '5509_Outdoor Temperature_TL Archive'
]])
```

Now we have categorize the data into 4 groups according to the time of operation, the load and the temperature. Let's visualize it.

```python
df_select3.boxplot(by='Class', column='5521_Building Load_TL Archive')
```

```python
df_select3.boxplot(by='Class', column='5509_Outdoor Temperature_TL Archive')
```

```python
df_select3.boxplot(by='Time', column='Class')
```

```python
df_select3.plot.scatter(
    x='5509_Outdoor Temperature_TL Archive', y='5521_Building Load_TL Archive', c='Class', colormap='viridis'
)
```

And we can conclude that within a month, the building's operation can be categorized according to the load and operation time in a day. Outdoor temperature? It's not an important factor in this case.
